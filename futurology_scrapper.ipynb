```python
from bs4 import BeautifulSoup
import requests
```

```python
url = 'https://futurology.life/101-most-innovative-cambridge-based-artificial-intelligence-companies/'
```

```python
page = requests.get(url)
soup = BeautifulSoup(page.text, 'html')
print(soup)
```

```python
# Find all the company links on the webpage
company_divs = soup.find_all("div", class_="wp-block-cover__inner-container")
```

```python
# Iterate over each company section and extract the desired information
for section in company_divs:
    # Extract company name
    company_name = section.find("h3").text.strip()
    
    # Extract website link
    website_link = section.find("a", href=lambda href: href and "http" in href)
    website_url = website_link["href"] if website_link else None
    
    # Extract LinkedIn link
    linkedin_link = section.find("a", href=lambda href: href and "linkedin.com" in href)
    linkedin_url = linkedin_link["href"] if linkedin_link else None
    
    # Extract Crunchbase link
    crunchbase_link = section.find("a", href=lambda href: href and "crunchbase.com" in href)
    crunchbase_url = crunchbase_link["href"] if crunchbase_link else None
    
    # Extract Twitter link
    twitter_link = section.find("a", href=lambda href: href and "twitter.com" in href)
    twitter_url = twitter_link["href"] if twitter_link else None
    
    # Extract brief description
    brief_description = section.find_all("p")[-1].text.strip()

    # Print the extracted information
    print("Company Name:", company_name)
    print("Website Link:", website_url)
    print("LinkedIn Link:", linkedin_url)
    print("Crunchbase Link:", crunchbase_url)
    print("Twitter Link:", twitter_url)
    print("Brief Description:", brief_description)
    print()
```
